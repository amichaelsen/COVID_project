```python
import pandas as pd
```

```python
#data source: https://data.cdc.gov/Vaccinations/COVID-19-Vaccinations-in-the-United-States-Jurisdi/unsk-b7fc
vaccine_data = pd.read_csv('../data/us_vaccination_data.csv')
```

```python
vaccine_data.loc[:,'Date'] = pd.to_datetime(vaccine_data['Date'])
vaccine_data.head(2)
```

```python
vaccine_data.columns
```

Administered	
Total number of administered vaccines based on state __where administered__

Total number of doses administered to people 12+ based on the jurisdiction __where recipient lives__

```python
# convert state codes to state names
# dict creds to: https://gist.github.com/rugbyprof/76575b470b6772ce8fa0c49e23931d97
states_codes_dict = {"AL":"Alabama","AK":"Alaska","AZ":"Arizona","AR":"Arkansas",
          "CA":"California","CO":"Colorado","CT":"Connecticut","DE":"Delaware",
          "FL":"Florida","GA":"Georgia","HI":"Hawaii","ID":"Idaho",
          "IL":"Illinois","IN":"Indiana","IA":"Iowa","KS":"Kansas",
          "KY":"Kentucky","LA":"Louisiana","ME":"Maine","MD":"Maryland",
          "MA":"Massachusetts","MI":"Michigan","MN":"Minnesota","MS":"Mississippi",
          "MO":"Missouri","MT":"Montana","NE":"Nebraska","NV":"Nevada",
          "NH":"New Hampshire","NJ":"New Jersey","NM":"New Mexico","NY":"New York",
          "NC":"North Carolina","ND":"North Dakota","OH":"Ohio","OK":"Oklahoma",
          "OR":"Oregon","PA":"Pennsylvania","RI":"Rhode Island","SC":"South Carolina",
          "SD":"South Dakota","TN":"Tennessee","TX":"Texas","UT":"Utah","VT":"Vermont",
          "VA":"Virginia","WA":"Washington","WV":"West Virginia","WI":"Wisconsin","WY":"Wyoming"}
states_codes_dict['DC'] = 'District of Columbia'
```

```python
vaccine_data.replace({'Location':states_codes_dict}, inplace=True)
```

```python
# restrict down to contiguous regions in the US
#vaccine_data['Location'].unique()
contiguous_US = ['Michigan','Wyoming','Maryland','Massachusetts','Rhode Island','Oregon',
                 'Pennsylvania','West Virginia','District of Columbia','Maine','Indiana',
                 'Nevada','Georgia','New Mexico','Colorado','Virginia','Kentucky','Minnesota',
                 'Arkansas','Connecticut','North Carolina','Montana','Oklahoma','Louisiana',
                 'Nebraska','Alabama','Florida','North Dakota','Delaware','Vermont','Idaho',
                 'Kansas','Wisconsin','New Jersey','Iowa','Missouri','New York','Utah',
                 'Washington','Arizona','Tennessee','Ohio','Texas','Mississippi','South Carolina',
                 'California','New Hampshire','South Dakota','Illinois']
cont_vaccine_data = vaccine_data[vaccine_data['Location'].isin(contiguous_US)]
cont_vaccine_data['Location'].nunique()
```

```python
ca_data = cont_vaccine_data[cont_vaccine_data['Location']=='California']
```

```python
ca_data[['Date','Location','Administered', 'Administered_12Plus', 'Administered_18Plus', 'Administered_65Plus']].sort_values(by='Date')
```

```python
ca_data['Administered_12_to_17'] = ca_data['Administered_12Plus'] - ca_data['Administered_18Plus']
ca_data['Administered_18_to_64'] = ca_data['Administered_18Plus'] - ca_data['Administered_65Plus']
```

```python
ca_data[['Date','Location', 'Administered_12_to_17',
        'Administered_18_to_64', 'Administered_65Plus']].sort_values(by='Date')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# very inconsistent in when the data for smaller categories was reported, as seen by the jumps. 
sns.lineplot(data=ca_data, x='Date', y='Administered_12Plus', label="12+")
sns.lineplot(data=ca_data, x='Date', y='Administered', label="admin");
sns.lineplot(data=ca_data, x='Date', y='Administered_18Plus', label="18+");
sns.lineplot(data=ca_data, x='Date', y='Administered_65Plus', label="65+");
plt.legend();
```

```python
temp_df = ca_data.copy()
temp_df['admin_minus_18plus'] = temp_df['Administered'] - temp_df['Administered_18Plus'] 
temp_df['admin_minus_65plus'] = temp_df['Administered'] - temp_df['Administered_65Plus']
temp_df['admin_minus_12plus'] = temp_df['Administered'] - temp_df['Administered_12Plus']
temp_df['recip_admin_minus_admin'] = temp_df['Recip_Administered'] - temp_df['Administered'] 
temp_df = temp_df[temp_df['Date'] <= pd.to_datetime('2021-05-01')]
```

```python
# very inconsistent in when the data for smaller categories was reported, as seen by the jumps. 
sns.lineplot(data=temp_df, x='Date', y='Administered', label="admin");
#sns.lineplot(data=temp_df, x='Date', y='Administered_12Plus', label="12+");
#sns.lineplot(data=temp_df, x='Date', y='admin_minus_12plus', label="all-12+");

sns.lineplot(data=temp_df, x='Date', y='Administered_18Plus', label="18+");
sns.lineplot(data=temp_df, x='Date', y='admin_minus_18plus', label="all-18+");

sns.lineplot(data=temp_df, x='Date', y='Administered_65Plus', label="65+");
sns.lineplot(data=temp_df, x='Date', y='admin_minus_65plus', label="all-65+");

plt.legend();
```
